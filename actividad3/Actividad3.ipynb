```python
#Importaci√≥n de pandas, numpy y matplot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cargando los archivos del 22 de Diciembre y 22 de Junio
df0 = pd.read_csv('dec22.txt', skiprows=[0,2], sep='\s+')
```

```python
df1 = pd.read_csv('jun22.txt', skiprows=[0,2], sep='\s+')
```

```python
#Asignar un dataframe
df = pd.DataFrame(df0)
```

```python
dg = pd.DataFrame(df1)
```

```python
#Preguntando por los tipos de datos
df.dtypes
```

```python
#Muestra de los primeros 5 renglones
df.head()
```

```python
#Variaci√≥n de la presi√≥n con la altura el 22 de Diciembre
df3 = df[['HGHT','PRES']]
plt.figure(); df3.plot(x='PRES');plt.legend(loc='best') 
plt.title("Variaci√≥n de la Presi√≥n Dec22")
plt.ylabel("Altura(m)")
plt.xlabel("Presi√≥n(Pa)")
plt.xlim(0,1000)
plt.grid(True)
plt.show()
```

```python
#Variaci√≥n de la presi√≥n con la altura el 22 de Junio
df4 = dg[['HGHT','PRES']]
plt.figure(); df4.plot(x='PRES');plt.legend(loc='best') 
plt.title("Variaci√≥n de la Presi√≥n Dec22")
plt.ylabel("Altura(m)")
plt.xlabel("Presi√≥n(Pa)")
plt.xlim(0,1000)
plt.grid(True)
plt.show()
```

```python
#Temperatura y Roc√≠o en funci√≥n de la altura 22 de Diciembre
df5 =df[['TEMP','DWPT']]
plt.figure(); df5.plot()
plt.plot(x=df.HGHT)
plt.title("Temperatura y Temp. del Roc√≠o seg√∫n la altura Dec22")
plt.xlabel("Altura(m)")
plt.ylabel("Temperatura y Temp. del Roc√≠o (¬∞C)")
plt.show()
```

```python
#Temperatura y Roc√≠o en funci√≥n de la altura 22 de Junio
df6 =dg[['TEMP','DWPT']]
plt.figure(); df6.plot()
plt.plot(x=dg.HGHT)
plt.title("Temperatura y Temp. del Roc√≠o seg√∫n la altura Jun22")
plt.xlabel("Altura(m)")
plt.ylabel("Temperatura y Temp. del Roc√≠o (¬∞C)")
plt.show()
```

```python
#Rapidez de los vientos en nudos 22 de Diciembre
plt.figure(); df.SKNT.plot()
plt.title("Variaci√≥n de la Rapidez de los Vientos Dec22")
plt.ylabel("Rapidez (m/s)")
plt.ylim(0,200)
plt.grid(True)
plt.show()
```

```python
#Rapidez de los vientos en nudos 22 de Junio
plt.figure(); dg.SKNT.plot()
plt.title("Variaci√≥n de la Rapidez de los Vientos Dec22")
plt.ylabel("Rapidez (m/s)")
plt.ylim(0,200)
plt.grid(True)
plt.show()
```
